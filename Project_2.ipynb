üß† Develop a program that:

Takes input for 5 items: name, price, quantity (as strings)

Convert all values using casting to int and float

Store data in lists and dictionaries

Add 10% tax, calculate subtotal, total, and average

Identify high-value items (price > 500) and store in a set, convert to frozenset

Use functions: calculate_bill(), apply_discount(), print_cart_summary()

Use loops to display item info with type() checks

Use random.choice() for a giveaway item

Use keyword module to verify if item name is a Python keyword

Use datetime to log purchase time and calendar to show current month



```python
import datetime
import calendar
import random
import keyword

# ---------- Functions ----------
def calculate_bill(price, quantity, tax_rate=0.10):
    subtotal = price * quantity
    tax = subtotal * tax_rate
    total = subtotal + tax
    return round(subtotal, 2), round(tax, 2), round(total, 2)

def apply_discount(total, threshold=1000, discount_rate=0.05):
    if total > threshold:
        discount = total * discount_rate
        return round(total - discount, 2)
    return total

def print_cart_summary(cart, high_value_set, giveaway_item):
    print("\nüõçÔ∏è Cart Summary:")
    for item in cart:
        print(f"\nItem: {item['name']}")
        print(f"Price: {item['price']} ({type(item['price'])})")
        print(f"Quantity: {item['quantity']} ({type(item['quantity'])})")
        print(f"Subtotal: {item['subtotal']}")
        print(f"Tax: {item['tax']}")
        print(f"Total: {item['total']}")
        if item['is_keyword']:
            print("‚ö†Ô∏è Item name is a Python keyword!")

    print("\nüíé High-value Items (frozenset):", high_value_set)
    print("\nüéÅ Giveaway Item:", giveaway_item)

# ---------- Main Program ----------
cart_items = []
high_value_items = set()

print("\nüïí Purchase Time:", datetime.datetime.now())
print("üìÖ Current Month:\n", calendar.month(datetime.datetime.now().year, datetime.datetime.now().month))

for i in range(5):
    print(f"\nEnter details for Item {i + 1}")
    name = input("Item name: ")
    price = float(input("Price: "))
    quantity = int(input("Quantity: "))

    subtotal, tax, total = calculate_bill(price, quantity)
    is_kw = keyword.iskeyword(name)

    item_dict = {
        'name': name,
        'price': price,
        'quantity': quantity,
        'subtotal': subtotal,
        'tax': tax,
        'total': total,
        'is_keyword': is_kw
    }

    if price > 500:
        high_value_items.add(name)

    cart_items.append(item_dict)

# Convert high-value items set to frozenset
high_value_frozenset = frozenset(high_value_items)

# Choose giveaway item randomly
giveaway = random.choice(cart_items)['name']

# Print summary
print_cart_summary(cart_items, high_value_frozenset, giveaway)

# Calculate final bill with optional discount
final_total = sum(item['total'] for item in cart_items)
final_total_with_discount = apply_discount(final_total)

print("\nüßæ Final Total (before discount):", round(final_total, 2))
print("üí∏ Final Total (after discount if applicable):", round(final_total_with_discount, 2))
print("\n‚úÖ Purchase Completed.")
```
