üß† Create a Python program where you:

Take input for 5 students (name, age, city, and 3 subject scores as strings)

Store data using lists, tuples, and dictionaries

Use type casting to convert scores to integers

Calculate total, average (using statistics.mean()), and grade using if-elif-else

Define functions: calculate_grade(), is_passed(), display_student_info()

Store results in a set, remove duplicates, and convert to frozenset

Use datetime to log current report time

Use math to find square root of total marks

Use keyword to check if names are Python keywords

Display data types using type()

Use os to print current working directory

Print final results using loops and conditions


```python
import statistics
import datetime
import math
import keyword
import os

# ---------- Functions ----------
def calculate_grade(avg):
    if avg >= 90:
        return 'A'
    elif avg >= 80:
        return 'B'
    elif avg >= 70:
        return 'C'
    elif avg >= 60:
        return 'D'
    else:
        return 'F'

def is_passed(scores):
    return all(score >= 40 for score in scores)

def display_student_info(student):
    print("\n--- Student Info ---")
    for key, value in student.items():
        print(f"{key.capitalize()}: {value}")
    print("Passed:", "Yes" if student['passed'] else "No")

# ---------- Main Program ----------
students_list = []
names_set = set()

print("\nüìå Report Generated On:", datetime.datetime.now())

for i in range(5):
    print(f"\nEnter details for Student {i + 1}")
    name = input("Name: ")
    age = input("Age: ")
    city = input("City: ")

    scores_input = input("Enter 3 subject scores separated by space: ")
    scores = [int(s) for s in scores_input.split()]

    # Tuple (optional)
    student_tuple = (name, age, city, scores)

    # Calculate total, average
    total = sum(scores)
    average = statistics.mean(scores)
    grade = calculate_grade(average)
    passed = is_passed(scores)
    sqrt_total = math.sqrt(total)  # ‚àötotal using math

    # Dictionary to hold data
    student_dict = {
        'name': name,
        'age': int(age),
        'city': city,
        'scores': scores,
        'total': total,
        'average': average,
        'grade': grade,
        'passed': passed,
        'sqrt_total': round(sqrt_total, 2),
        'is_keyword': keyword.iskeyword(name),
        'types': {
            'name': type(name),
            'age': type(int(age)),
            'city': type(city),
            'scores': type(scores),
            'average': type(average)
        }
    }

    students_list.append(student_dict)
    names_set.add(name)

# Remove duplicates and convert to frozenset
unique_names = frozenset(names_set)

# ---------- Output ----------
print("\nüìÇ Current Working Directory:", os.getcwd())
print("\nüßä Unique Names (frozenset):", unique_names)

for student in students_list:
    display_student_info(student)

    print("‚úî Data Types:")
    for field, dtype in student['types'].items():
        print(f"  - {field}: {dtype}")

    if student['is_keyword']:
        print("‚ö†Ô∏è Note: Name is a Python keyword!")
    else:
        print("‚úÖ Name is valid (not a Python keyword)")

# ---------- End ----------
print("\n‚úÖ Report Completed.")
```
